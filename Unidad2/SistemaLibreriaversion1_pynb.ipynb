```python
##Definicion de clases
class Libro:
  def __init__(self,isbn, titulo, autor,categoria,precio):
    self.isbn=isbn
    self.titulo=titulo
    self.autor=autor
    self.categoria=categoria
    self.precio=precio
    self.prestado = False

  def mostra_info(self):
    estado="Prestado" if self.prestado else "Disponible"
    print(f"[{self.isbn}]{self.titulo} - {self.autor} | {self.categoria} | ${self.precio} | {estado}")

class Estudiante:
  def __init__(self,num_control,nombre,carrera,fecha_nacimineto,telefono,correo):
    self.num_control=num_control
    self.nombre=nombre
    self.carrera=carrera
    self.fecha_nacimineto=fecha_nacimineto
    self.telefono=telefono
    self.correo=correo

  def mostra_info(self):
    print(f"[{self.num_control}],{self.nombre},-{self.carrera}|3{self.fecha_nacimineto}|\
      {self.telefono}|{self.correo}")

class Prestamo:
  contadores_prestmao=1

  def __init__(self,num_control,isbn,fecha_inicio,fecha_terminio):
    self.folio = Prestamo.contadores_prestmao
    Prestamo.contadores_prestmao+=1
    self.num_control=num_control
    self.isbn=isbn
    self.fecha_inicio=fecha_inicio
    self.fecha_terminio=fecha_terminio

  def mostrar_info(self):
    print(f"Folio{self.folio}| Estudiante:{self.num_control}|Libro:{self.isbn}\
      Del{self.fecha_inicio} al {self.fecha_terminio}")

class Venta:
  contadores_venta=1

  def __init__(self,num_control,isbn,fecha,cantidad,importe):
    self.folio_venta=Venta.contadores_venta
    Venta.contadores_venta +=1
    self.isbn=isbn
    self.num_control=num_control
    self.fecha=fecha
    self.cantidad=cantidad
    self.importe=importe

  def mostras__info(self):
    print(f"Folio:{self.folio_venta}|ISBN{self.isbn}|Comprador{self.num_control}|\
      Fecha:{self.fecha}|Cantidad:{self.cantidad}|Importe $:{self.importe}")

###tarea hacer la funcion de alta libro
def alta_Libro():
  print("\n----Alta libro----")
  isbn=input("ISBN:")
  for libro in libros:
    if libro.isbn==isbn:
      print("Ya existe un libro con ese ISBN")
      return
  titulo=input("Titulo:")
  autor=input("Autor:")
  categoria=input("Categoria:")
  try:
    precio=float(input("Precio:"))
  except ValueError:
    print("Precio invalido")
    return
  Nuevo_libro=Libro(isbn,titulo,autor,categoria,precio)
  libros.append(Nuevo_libro)
  print("Libro registrado exitosamente")

###Consultar los libros
def consultsr_libro():
  print("\n----Consultar libro----")
  isbn=input("ISBN:")
  for libro in libros:
    if libro.isbn==isbn:
      libro.mostra_info()
      return
  print("No existe un libro con ese ISBN")

###Mostrar todosl os libros
def mostrar_todos_libros():
  print("\n----Mostrar todos los libros----")
  if not libros:
    print("No hay+libros registrados")
  else:
    for libro in libros:
      libro.mostra_info()

###Consultar los estudiantes
def consultar_estudiante():
  print("\n----Consultar estudiante----")
  num_control=input("Numero de control:")
  for est in estudiantes:
    if est.num_control==num_control:
      est.mostra_info()
      return
  print("Estudiante no encontrado. ")

##Mostrar todos los estudiantes
def mostrar_todos_estudiantes():
  print("\n----Mostrar todos los estudiantes----")
  if not estudiantes:
    print("No hay estudiantes registrados")
  else:
    for est in estudiantes:
      est.mostra_info()


##Dar de alta el estudiante
def alta_estudiante():
  print("\n----Alta estudiante----")
  num_control=input("Numero de control:")
  for est in estudiantes:
    if est.num_control==num_control:
      print("Ya existe un estudiante con ese numero de control")
      return
  nombre=input("Nombre:")
  carrera=input("Carrera:")
  fecha_nacimiento=input("Fecha de nacimiento:")
  telefono=input("Telefono:")
  correo=input("Correo:")
  nuevo_estudiante=Estudiante(num_control,nombre,carrera,fecha_nacimiento,telefono,correo)
  estudiantes.append(nuevo_estudiante)
  print("Estudiante registrado exitosamente")

##Realizar prestamo
def realizar_prestamo():
  print("\n----Realizar prestamo----")
  num_control=input("Numero decontrol del estudiante:")
  isbn=input("ISBN del libro:")
  found_estudiantes=[est for est in estudiantes if est.num_control==num_control]
  if not found_estudiantes:
    print("Estudiante no encontrado")
    return
  found_libros=[libro for libro in libros if libro.isbn==isbn]
  if not found_libros:
    print("Libro no encontrado")
    return
  fecha_inicio=input("Fecha de inicio (dd/mm/aaaa):")
  fecha_terminio=input("Fecha de terminio (dd/mm/aaaa):")
  new_prestamo=Prestamo(num_control,isbn,fecha_inicio,fecha_terminio)
  prestamos.append(new_prestamo)
  found_libros[0].prestado=True
  print(f"Prestamo registrado ")

##Mostrar prestamo
def mostrar_prestamos():
  print("\n----Mostrar prestamos----")
  if not prestamos:
    print("No hay prestamos registrados")
  else:
    for prestamo in prestamos:
      prestamo.mostrar_info()

##devolver el libro
def devolver_libro():
  print("\n----Devolver libro----")
  try:
    folio=input("Folio del prestamo a devolver :")
  except ValueError:
      print("Folio invalido")
      return
  prestamo=next((prestamo for prestamo in prestamos if prestamo.folio==folio),None)
  if not prestamo:
    print("Prestamo no encontrado")
    return
  libro=next((libro for libro in libros if libro.isbn==prestamo.isbn),None)
  if libro:
    libro.prestado=False
    prestamos.remove(prestamo)
    print(f"Libro {libro.titulo} devuelto correctamente")
  else:
    print("No se encontro el libro prestado")

##registros de ventas
def registrar_ventas():
  print("\n----Registrar venta----")
  isbn=input("ISBN del libro:")
  libro=next((libro for libro in libros if libro.isbn==isbn),None)
  if not libro:
    print("Libro no encontrado")
    return
  try:
    cantidad=int(input("Cantidad a vender (solo 1):"))
    if cantidad !=1:
      print("Solo se puede vender un libro por vez")
      return
  except ValueError:
    print("Cantidad invalida")
    return
  num_control=input("Numero de control del comprador (999999 si es publico general):")
  if num_control!="999999":
    estudiante=next((est for est in estudiantes if est.num_control==num_control),None)
    if not estudiante:
      print("Estudiante no encontrado")
      return
  fecha:input("Fecha de la venta (dd/mm/aaaa):")
  importe=libro.precio*cantidad
  venta=Venta(isbn,num_control,fecha,cantidad,importe)
  ventas.append(venta)
  libros.remove(libro)
  print(f"Venta registrada.  Folio{venta.folio_venta}")

##Mostrar historial de ventas
def mostrar_historial_ventas():
  print("\n----Mostrar historial de ventas----")
  if not ventas:
    print("No hay ventas registradas")
  else:
    for v in ventas:
      v.mostrar_info()



###funcion de menu}
def mostrar_Menu():
  print("\n---SISTEMA DE LIBRWERIS ESCOLAR")
  print("1. Dar de alta de libro")
  print("2. Dar de alta de estudiante")
  print("3.consultar libro por ISBN")
  print("4.consultar estudiante por numero de control")
  print("5.Mostrar todos los libros")
  print("6.Mostrar todos los estudiantes")
  print("7.Realizar prestamo")
  print("8.Mostrar prestamo")
  print("9.Devolver libro")
  print("10.Realizar ventas")
  print("11.Mostar historial de ventas")
  print("12.Alir del sistema")
  return input("Seleccione una opcion: ")

def iniciar_Systema():
   while True:
     opcion = mostrar_Menu()
     if opcion == "1":
       alta_Libro()
     elif opcion == "2":
       alta_estudiante()
     elif opcion=="3":
       consultsr_libro()
     elif opcion=="4":
       consultar_estudiante()
     elif opcion=="5":
       mostrar_todos_libros()
     elif opcion=="6":
       mostrar_todos_estudiantes()
     elif opcion=="7":
       realizar_prestamo()
     elif opcion=="8":
       mostrar_prestamos()
     elif opcion=="9":
      devolver_libro()
     elif opcion=="10":
        registrar_ventas()
     elif opcion=="11":
      mostrar_historial_ventas()
     elif opcion=="12":
       print("Saliendo del sistema, ¡hasta luego!")
       break
     else:
       print("Opcion no valida. Intente de nuevo")

###Programa principal###

#Iniciar sistema
libros=[]
estudiantes=[]
prestamos=[]
ventas=[]
iniciar_Systema()
```
